# Questão 02:

Considere a seguinte função de transferência:

$ G(s)= \frac{1}{s^2 + 0.8s + 1}$

Utilizando a linguagem de programação Python calcular o pico de ressonância em $dB$ e freqüência em que ocorre o pico de ressonância em $\frac{rad}{s}$.



# Solução da Questão 2

```
!pip install control
```

```
import numpy as np
import matplotlib.pyplot as plt
from control import *
num = [1]
den = [1, 0.8, 1]
sistema = tf(num,den)

w = np.logspace(-2,2,2000)

[re,im,w] = nyquist(sistema,w)            #left hand arguments                                   
relacao = (re**2 + im**2)**(0.5)          #relação para cálculo do vetor girante a partir do gráfico de Nyquist
modulo = max((re**2 + im**2)**(0.5))
modulodb = 20*np.log10(modulo)

i=0
for x in relacao:
  i=i+1
  if x == modulo:                         #parar quando o x for o pico de ressonancia
   break 

print("O pico de ressonância em dB é:")
print (modulodb)
print("freqüência [rad/s] que ocorre o pico de ressonância é")
print(w[i])                                #mostra a frequencia relacionada ao ponto onde esta o pico


```

### Interpretação da curva para a determinação da FT:

Observa-se que a curva começa decrescendo -20dB/dec, logo depois de w = 1 rad/se mantém constante com o valor de 12dB. Após w = 10 rad/s, a curva volta a decair, mas dessa vez a -40dB/dec.

Tal comportamento representa numericamente:


*   Um polo na origem (1/s), contribuindo com -90º
*   Para que o gráfico fique paralelo ao eixo real de 1 a 10 rad/s, há um fator derivativo (s+1) para "anular" o decaimento inicial de -20dB/dec.Contribuiçao de 45º.
   
*   O decaimento de -40dB/dec revela a presença de polos complexos conjugados, o que se traduz em: 1/(0,1s + 1)^2. Contribuição de -180º
*   O ganho K é definido com a reta horizontal em 12dB, logo 20log|20| = 12, logo K = 3,981. Contribuição de 0º.



*   Junrando todas as contribuições, encontra-se a Função de Transferência G(s), tal que:

G(s) = (1/s) * (s+1) * 1/(0,1s + 1)^2 * 3,981

$G(jw)$ = $3,981(s+1)\over s(0,1s+1)^2$
