# Começando a trabalhar com Teoria de Controle

## Exemplo 7.1 (Ogata)

Sistema de Primeira Ordem 

\begin{equation}
G(s) = \frac{K}{Ts+1}
\end{equation}

Módulo e ângulo da função de transferência senoidal

\begin{equation}
\left | G(j \omega) \right | = \frac{K}{\sqrt{1^2 + \omega^2*T^2}} 
\end{equation}

\begin{equation}
\sphericalangle G(j \omega) = - atan(T\omega)
\end{equation}


```
import numpy as np
from plotly import graph_objects as go 
from plotly.subplots import make_subplots

T = 1

freq = np.logspace(-1,3,10)

modulo = 1/np.sqrt(1+freq**2*T**2)
angulo = -np.arctan(freq*T)*180/np.pi

figura = make_subplots(rows=2,cols=1)

figura.add_trace(go.Scatter(y=modulo,x=freq,name="Módulo"), row=1,col=1)
figura.add_trace(go.Scatter(y=angulo,x=freq,name="Ângulo"), row=2,col=1)

figura.update_xaxes(type="log")

figura.show()
```

## Alguns exemplos de utilização da biblioteca "control"

```
from control.matlab import *
import numpy as np
from plotly import graph_objects as go

num = [1, 3]
den = [2, 3, 8]

gs = tf(num, den)

K=1
T=1
num = [K]
den = [T, 1]

gs= tf(num,den)

t = np.linspace(0,5,10000)
u = 10*np.sin(10*t)
x0 = 0

saida = lsim(gs,u,t,x0)

figura = go.Figure(data=go.Scatter(y=saida[0],x=t,name="Saída"))
figura.add_trace(go.Scatter(y=u,x=t,name="Entrada"))

figura.show()
```
